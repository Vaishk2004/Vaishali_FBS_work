```python
import pandas as pd
```

```python
df = pd.read_csv("cars93.csv")
```

```python
df.head()
```

```python
df['Type'].unique()
```

```python
from sklearn.preprocessing  import LabelEncoder
le = LabelEncoder()
```

```python
df['Type'] = le.fit_transform(df['Type'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Compact','Small','Midsize','Large','Sporty','Van']])
```

```python
df['Type'] = oe.fit_transform(df[['Type']]).astype('int')
```

```python
df.head()
```

```python
df['Type'].head()
```

One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
df.dropna(inplace = True)
```

```python
result = ohe.fit_transform(df[['AirBags']])    #Gives output in the form of matrix
```

```python
result = result.toarray()
```

```python
result = pd.DataFrame(result,columns = ['Driver & Passenger','Driver only'])
```

```python
#result
```

```python
df = pd.concat([df,result],axis=1)
```

```python
df.head()
```

```python
df.drop('AirBags',axis=1,inplace=True)
df.head()
```
